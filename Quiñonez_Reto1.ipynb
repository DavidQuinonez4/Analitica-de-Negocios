En este trabajo se analiza el uso del modelo Naive Bayes en una entidad financiera como apoyo a los procesos de prevención del lavado de activos. A partir del análisis de datos, el objetivo es comprender mejor el comportamiento financiero de las personas y contribuir a la detección temprana de posibles riesgos dentro del sistema SARLAFT. Para ello, se utilizan distintas variables que permiten describir tanto aspectos socioeconómicos como financieros de los individuos evaluados. Las variables que se tendrán en cuenta son:

**Edad:** es la edad que posee el empleado.

**Ingresos:** Ingresos recibidos por una personas como resultado de una actividad comercial.

**Egresos:** Son los gastos mensuales de una persona, gastos de una persona a cargo, hijos y servicios públicos.

**Número de tarjetas de crédito:** Corresponde a la cantidad de tarjetas de crédito activas que tiene una persona, lo cual permite analizar su nivel de uso y acceso a productos financieros.

**Monto transado con tarjetas (USD):** Representa el valor acumulado de las transacciones realizadas mediante tarjetas de crédito, expresado en dólares estadounidenses, y sirve como referencia del nivel de actividad financiera.

**Porcentaje de crecimiento del patrimonio:** Hace referencia al cambio porcentual del patrimonio de una persona durante un periodo determinado, información útil para identificar variaciones inusuales en su situación económica.

**Lavado de activos:** Se refiere a las actividades ilegales orientadas a encubrir el origen real de recursos o bienes, con el fin de hacerlos pasar como legítimos.

0. Cargamos las librerías

```python
import numpy as np #Librería numérica por excelencia.
import pandas as pd #Librería para la comunicación con archivos de excel

#Librerías específicas
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Cargamos la Base de Datos

```python
nxl = "/content/2. LavadoActivos.xlsx"
XDB = pd.read_excel(nxl)
XDB.head()

XD=XDB[["edad","ingresos_usd", "gastos_usd", "num_tarjetas_credito","monto_transado_tarjetas_usd","porcentaje_crecimiento_patrimonio"]] #Variables de trabajo
XD.head(10)
yd=XDB[["lavado_activos"]]
yd.head()
```

2. Implementamos modelo Naive Bayes

```python
np.set_printoptions(suppress=True,precision=3)
mnb=GaussianNB()
mnb.fit(XD,yd)

u=mnb.theta_
sigma=mnb.var_;sigma=np.sqrt(sigma)
print("edad","ingresos_usd", "gastos_usd", "num_tarjetas_credito","monto_transado_tarjetas_usd","porcentaje_crecimiento_patrimonio", "lavado_activos")
print ("Las medias son:")
print(u)
print ("Las desviaciones son:")
print(sigma)
print('Los valores maximos son:\n',u+sigma)
print('Los valores minimos son:\n',u-sigma)
```

**Análisis de Resultados**

Los resultados obtenidos a partir del modelo Naive Bayes muestran una diferencia marcada entre los individuos clasificados como posibles casos de lavado de activos y aquellos que no presentan este riesgo. En particular, las variables asociadas al manejo de recursos financieros, como los ingresos, los gastos y el monto total transado con tarjetas de crédito, presentan valores promedio considerablemente más altos en la clase correspondiente al lavado de activos. Esto indica que el modelo identifica principalmente el volumen de movimiento financiero como un factor clave en la clasificación.

Asimismo, se observa que dentro del grupo asociado al lavado de activos existe una mayor dispersión en los valores de ingresos y gastos, lo que evidencia la presencia de distintos perfiles financieros y comportamientos económicos variados. Por el contrario, los individuos que no presentan riesgo de lavado muestran un comportamiento más estable y homogéneo en estas variables.

En cuanto a la edad y al número de tarjetas de crédito, los promedios son muy similares entre ambas clases, lo que sugiere que estas variables no tienen un peso significativo en la diferenciación realizada por el modelo y cumplen principalmente un papel contextual. De manera similar, el porcentaje de crecimiento del patrimonio no presenta diferencias relevantes entre los grupos analizados, por lo que su influencia en la clasificación es limitada. En conjunto, los resultados confirman que el modelo se apoya principalmente en variables monetarias para identificar posibles riesgos de lavado de activos, lo cual es coherente con los enfoques utilizados en los sistemas de control SARLAFT.
